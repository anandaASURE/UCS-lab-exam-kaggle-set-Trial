```python
import pandas as pd
import numpy as np
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

### now check data is there or not 

```python
train_df.head(4)
```

```python
test_df.head(4)
```

## splitting NOW the features colums in one var and target in other 

```python
x=train_df.drop(columns=['target'])
y=train_df['target']
x.head(2)
```

```python
y.head(3)
```

# NOW WE HAVE TO SCALE OUR FEATURES

```python
scl=StandardScaler()
x_scaled= scl.fit_transform(x)
x_scaled[:5]
```

# Split into training & validation sets

```python
x_train, x_val, y_train, y_val = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
x_train[:3]
```

```python
x_val[:3]
```

```python
!pip install --upgrade xgboost
```

```python
xgb_model = xgb.XGBRegressor(n_estimators=500, learning_rate=0.05, max_depth=6, 
                             subsample=0.8, colsample_bytree=0.8, random_state=42)
xgb_model.fit(x_train, y_train, eval_set=[(x_val, y_val)], verbose=True)
```

# Evaluate on validation set

```python
y_pred = xgb_model.predict(x_val)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f"Validation RMSE: {rmse}")
print("Validation Predictions Head:\n", y_pred[:5])
```

# Prepare test data & make predictions

```python
x_test = scl.transform(test_df.drop(columns=["id"]))
print("Test Features Scaled Sample:\n", x_test[:5])
test_predictions = xgb_model.predict(x_test)
print("Test Predictions Head:\n", test_predictions[:5])
```

## now submit

```python
submission = pd.DataFrame({"id": test_df["id"], "target": test_predictions})
print("Submission Data Head:\n", submission.head())
submission.to_csv("submission.csv", index=False)
print("Submission file saved.")
```
